```
def mean(x):
  return sum(x)/ len(x)

def median(v):
  
  n = len(v)
  sorted_v = sorted(v)
  median_point = n // 2
  
  if n %2 == 1:
    return sorted_v[median_point]
  else:
    lo_point = sorted_v[median_point-1]
    hi_point = sorted_v[median_point]
    return (lo_point + hi_point) /2


def quantile(x, p):
  p_index = int(p * len(x))
  return sorted(x)[p_index]

def mode(x):
  counts = Counter(x)
  max_count = max(counts.values())
  return [x_i for x_i, count in counts.items() if count == max_count]# in python3 removed iteritems()

def data_range(x):

  return max(x)-min(x)

print(mean(num_friends))
print(median(num_friends))
print(quantile(num_friends, 0.10))
print(quantile(num_friends, 0.25))
print(quantile(num_friends, 0.75))
print(quantile(num_friends, 0.90))
print(mode(num_friends))
print(data_range(num_friends))
```
